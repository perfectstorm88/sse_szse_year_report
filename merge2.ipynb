```python
dfs = []
# for f in os.listdir():
#     if f.endswith('xlsx'): 
#         dfs += [pd.read_excel(f,index_col=None)]
dfs.append(pd.read_excel('./out/研究数据0408.xlsx',index_col=None))
dfs.append(pd.read_excel('./ROA数据来源0408_去重.xlsx',index_col=None))
dfs.append(pd.read_excel('./托宾Q值来源0408.xlsx',index_col=None))
len(dfs)
```

```python
print(dfs[2].shape)
dfs[2].head()
```

```python
dfs[1] = dfs[1].rename(columns={'Stkcd':'Symbol','Accper':'EndDate'})
dfs[1].loc[0]['Symbol']="证券代码'"
dfs[1].loc[0]['EndDate']="统计截止日期'"
dfs[1].head()
```

```python
dfs[1] = dfs[1][['Symbol','EndDate', 'F050201B', 'F050202B','F050203B','F050204C']]
dfs[1].head()
```

```python
dfs[2] = dfs[2].rename(columns={'Stkcd':'Symbol','Accper':'EndDate'})
dfs[2].head()
```

```python
dfs[2] = dfs[2][['Symbol','EndDate', 'F100501A', 'F100801A','F100802A','F100901A','F100902A','F100903A','F100904A','备注']]
dfs[2].loc[0]['Symbol']="证券代码'"
dfs[2].loc[0]['EndDate']="统计截止日期'"
dfs[2].head()
```

```python
max_row, total_column=max([v.shape[0] for v in dfs]),sum([v.shape[1] for v in dfs])
max_row,total_column
```

```python
df = dfs[0].merge(dfs[1],how='left',on=['Symbol','EndDate']).merge(dfs[2],how='left',on=['Symbol','EndDate'])
print(df.shape)
df.head()
```
