```python
# 民营上市公司数据 
# 国有股持股数 
# 研发投入占比.xlsx
# 资产负债和利润.xlsx
df = dfs[1].merge(dfs[0],how='outer').merge(dfs[2],how='outer').merge(dfs[3],how='outer')
print(df.shape)
df = df.dropna(subset=['Conme']) #过滤证券简称和中文名为空的
print(df.shape)
df.head()
```

```python
# 过滤 没有包含2015 2016 2017 这三年数据的公司
idMap = {}
for a in df.to_dict('records'):
    id = a['Stkcd']
    year = a['Reptdt']
    if not a['Stknme']:
        continue
    if '2018' in year:
        continue

    if id not in idMap:
        idMap[id] = [];
    idMap[id].append(year[0:4])

reversedId = []
expected = set(['2015','2016','2017'])
for (k,v) in  idMap.items(): 
    if set(v) & expected:
        reversedId.append(k)
    
print(len(reversedId),len(idMap.keys())) 
df2 = df[df["Stkcd"].isin(reversedId)]  
print(df.shape)
print(df2.shape)
# 还剩下 3513记录
```

```python
# 过滤 名称中包含ST的
import functools 
idMap = {}
for a in df.to_dict('records'):
    id = a['Stkcd']
    if id not in idMap:
        idMap[id] = [];
    idMap[id].append(a['Stknme'])
reversedId = []
for (k,v) in  idMap.items(): 
    stNum = functools.reduce(lambda x,y: (1+x) if 'ST' in y  else  (0 + x) , v, 0)
    if stNum <10:
        reversedId.append(k)
        

df3 = df2[df2["Stkcd"].isin(reversedId)]  
print(df2.shape)
print(df3.shape)
print(df2.shape[0]-df3.shape[0])  
    
```
