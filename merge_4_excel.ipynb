```python
"""
1.把下面四个文件合并为一个文件：
    国有股持股数.xlsx
    民营上市公司数据.xlsx
    研发投入占比.xlsx
    资产负债和利润.xlsx
    
2.过滤近三年缺失的记录

3.过滤10年ST的公司
"""
```

```python
import os
import pandas as pd
```

```python
dfs = []
for f in sorted(os.listdir()):
    if f.endswith('xlsx'): 
        print(f)
        dfs += [pd.read_excel(f,index_col=None)]
len(dfs)
```

```python
print(dfs[1].shape)
dfs[1].head()
```

```python
print(dfs[2].shape)
dfs[2].head()
```

```python
print(dfs[3].shape)
dfs[3].head()
```

```python
dfs[2] = dfs[2].rename(columns={'Symbol':'Stkcd','EndDate':'Reptdt'})
dfs[2].head()
```

```python
dfs[3] = dfs[3].rename(columns={'Symbol':'Stkcd','EndDate':'Reptdt'})
dfs[3].head()
```

```python
max_row, total_column=max([v.shape[0] for v in dfs]),sum([v.shape[1] for v in dfs])
max_row,total_column
```

```python
# 民营上市公司数据 
# 国有股持股数 
# 研发投入占比.xlsx
# 资产负债和利润.xlsx
df = dfs[1].merge(dfs[0],how='outer').merge(dfs[2],how='outer').merge(dfs[3],how='outer')
print(df.shape)
df = df.sort_values(by=['Stkcd', 'Reptdt'])
# 过滤2018
print('总公司数',len(set(df["Stkcd"])),len(df["Stkcd"]))    
df[df.apply(lambda x: '2018' not in x['Reptdt'],axis=1)]
print('过滤2018年数据后，总公司数',len(set(df["Stkcd"])),len(df["Stkcd"]))    
# df = df.dropna(subset=['Conme']) #过滤证券简称和中文名为空的
print(df.shape)
df.head()
```

```python
# import datetime
# date = datetime.datetime.now().strftime('%Y%m%d')
# df.to_excel('out_join_'+date+'.xlsx')

```

```python

x = df.to_dict('records')
x[0]
```

```python


# 过滤 没有包含2015 2016 2017 这三年数据的公司
idWithYear={}
for d in x:
    _id,year = d['Stkcd'],d['Reptdt'][0:4]
    if _id not in idWithYear:
        idWithYear[_id]=[]
    if year in ['2016','2017','2015']:
        idWithYear[_id].append(year)
deletedList = []
for (k,v) in  idWithYear.items(): 
    if len(v)<3:
        deletedList.append(k)
print('总公司数',len(set(df["Stkcd"])),len(df["Stkcd"]))      
print('没有包含2015 2016 2017 这三年数据的公司的公司数',len(deletedList)) 
df_haslast3year = df[~df["Stkcd"].isin(deletedList)]  
# print(df.shape)
# print(df2.shape)
# # 还剩下 3513记录
print('剩余公司数',len(set(df_haslast3year["Stkcd"])),len(df_haslast3year["Stkcd"])) 
```

```python
# 过滤 10年中名称中包含ST的
import functools 
idMap = {}
for a in df_haslast3year.to_dict('records'):
    id = a['Stkcd']
    if id not in idMap:
        idMap[id] = [];
    idMap[id].append(a['Stknme'])
reversedId = []
for (k,v) in  idMap.items(): 
    stNum = functools.reduce(lambda x,y: (1+x) if isinstance(y,str) and 'ST' in y  else  (0 + x) , v, 0)
    if stNum <10:
        reversedId.append(k)
        

df3 = df_haslast3year[df_haslast3year["Stkcd"].isin(reversedId)]  
print('过滤 10年中名称中包含ST的，剩余公司数',len(set(df3["Stkcd"])),len(df3["Stkcd"])) 
```

```python
from datetime import datetime
today = datetime.now().strftime("%Y%m%d")

df3.to_excel(f'./out/{today}_过滤ST_过滤近三年.xlsx')
```
