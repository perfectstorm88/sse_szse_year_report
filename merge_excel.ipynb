```python


# 过滤 没有包含2015 2016 2017 这三年数据的公司
idWithYear={}
for d in x:
    _id,year = d['Symbol'],d['EndDate'][0:4]
    if '证券代码' in _id:
        continue
    if _id not in idWithYear:
        idWithYear[_id]=[]
    if year in ['2016','2017','2015']:
        idWithYear[_id].append(year)
deletedList = []
for (k,v) in  idWithYear.items(): 
    if len(v)<3:
        deletedList.append(k)
print('总公司数',len(set(df["Symbol"])),len(df["Symbol"]))      
print('没有包含2015 2016 2017 这三年数据的公司的公司数',len(deletedList)) 
display(df.head())
df_haslast3year = df[~df["Symbol"].isin(deletedList)]  
# print(df.shape)
# print(df2.shape)
# # 还剩下 3513记录
print('剩余公司数',len(set(df_haslast3year["Symbol"])),len(df_haslast3year["Symbol"])) 
df_haslast3year.head()
```

```python
# 过滤 10年中名称中包含ST的
import functools 
idMap = {}
for a in df_haslast3year.to_dict('records'):
    id = a['Symbol']
    if id not in idMap:
        idMap[id] = [];
    idMap[id].append(a['ShortName'])
reversedId = []
for (k,v) in  idMap.items(): 
    stNum = functools.reduce(lambda x,y: (1+x) if isinstance(y,str) and 'ST' in y  else  (0 + x) , v, 0)
    if stNum <10:
        reversedId.append(k)
        

df3 = df_haslast3year[df_haslast3year["Symbol"].isin(reversedId)]  
print('过滤 10年中名称中包含ST的，剩余公司数',len(set(df3["Symbol"])),len(df3["Symbol"])) 
df=df3
```
