```python
import os
import pandas as pd
df1 = pd.read_excel('研究数据0423.xlsx',index_col=None)
df2 = pd.read_excel('上市公司基本信息年度表2000-2017.xlsx',index_col=None)
```

```python
df1.head()
```

```python
df1_=df1.drop(columns=['Unnamed: 0']) #删除第一列
df1_=df1_.drop([0]) #删除第一行
df1_.head()
```

```python
df2.head(3)
```

```python
df2_=df2.drop([1])
df2_.loc[0,'EndDate']="统计截止日期'"
df2_.loc[0,'Symbol']="证券代码'"

df2_.head(3)
```

```python
df2_=df2_[['Symbol','EndDate','ShortName','IndustryName','IndustryCode','RegisterAddress','Zipcode','EstablishDate']]
df2_.loc[:,"EstablishDate"] =[ i[:4] if type(i) is str and len(i)>9 else i for i in df2_["EstablishDate"]]
df2_.head()
```

```python
list1 = df1_.columns.tolist()
list2=['Stknme','Nindcd','Area','Listdt']
clist = [i for i in list1 if i not in list2] 
clist
df1_ = df1_[clist]
df1_.head()
```

```python
df1_ret =df1_.merge(df2_,how='left',on=['Symbol','EndDate'])
df1_ret.head()
```

```python
print(df2_.columns.tolist())
last_columns = clist[:2]+df2_.columns.tolist()[2:]+clist[2:]
print(last_columns)
df1_ret = df1_ret[last_columns]
df1_ret.head()
```
